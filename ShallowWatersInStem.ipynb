```python
"""
Make a pie chart - see
http://matplotlib.sf.net/matplotlib.pylab.html#-pie for the docstring.

This example shows a basic pie chart with labels optional features,
like autolabeling the percentage, offsetting a slice with "explode",
adding a shadow, and changing the starting angle.

"""
from pylab import *
colors = ["SteelBlue",
    "IndianRed","gold",
    
          "#eeeeee",
    "MediumTurquoise",
    "DarkOrange",
    "#E24A33"]

# make a square figure and axes
figure(1, figsize=(15,15))
ax = axes([0.1, 0.1, 0.8, 0.8])

# The slices will be ordered and plotted counter-clockwise.
labels = 'African American', 'Asian', 'Hispanic', 'White', 'other'
fracs = [2.1,14.3,3.2,79.2, 1.2]
explode=( 0.3, 0,0.3, 0, 0)
fracs2 = [6,10,5,78,1]
fracs3 = [12.2,4.7,16.3,63.7,2]

pie(fracs, explode=explode, labels=labels, 
    
                autopct='%1.1f%%', shadow=True, startangle=0, colors=colors[:len(fracs)])
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.
```

```python
figure(1, figsize=(15,15))
ax = axes([0.1, 0.1, 0.8, 0.8])

pie(fracs2, explode=explode, labels=labels,colors=colors,
                autopct='%1.1f%%', shadow=True, startangle=0)
                # The default startangle is 0, which would start
                # the Frogs slice on the x-axis.  With startangle=90,
                # everything is rotated counter-clockwise by 90 degrees,
                # so the plotting starts on the positive y-axis.


                
```

```python
def plotts(i,ax):
    pylab.rcParams.update(s)
    
    ks= [float(k) for k in gradstud.columns[1:]]
    #print  gradstud['2000'][cs::]
    #print [k for k in gradstud[:][cs]]
    ax.plot(ks[:10],[float(gradstud[k][cs+i*43].replace(',','')) for k in gradstud.columns[1:][:10]], label='CS', lw=3)
    ax.plot(ks[:10],[float(gradstud[k][astro+i*43].replace(',','')) for k in gradstud.columns[1:][:10]], label='Astro', lw=3)
    ax.plot(ks[:10],[float(gradstud[k][physical+i*43].replace(',','')) for k in gradstud.columns[1:][:10]], label='Physics', lw=3)

    pl.xticks(range(2000,2010),[k for k in gradstud.columns[1:]][:10])
    pl.xticks(rotation=90)
    #pl.plot(ks,[float(gradstud[k][cs]) for k in gradstud.columns[1:]])
    ax.set_ylim(pl.ylim()[0], pl.ylim()[1]*1.3)
    ax.legend(loc=2, ncol=3)
```

```python
for f in classes:
    fig=pl.figure(figsize=(10,6))
    ax=fig.add_subplot(111)
    ax.set_title(f)
    plotts(classes[f],ax)
```
